```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df= pd.read_csv("/content/wdi_data (1).csv")
df.head()
```

```python
 df = df[df["indicator"]== "SP.POP.TOTL"]

 df.head(20)
```

```python
 df = df[df["Country Code"] == "USA"]
 df.head()
```

```python
fig= go.Figure()

trace_0 = go.Bar(
    x=df["Year"],
    y=df["value"]
)

fig.add_trace(trace_0)

fig.show()
```

```python
fig.update_layout(
    title="USA Population Growth",
    xaxis= {"title":"Year"},
    yaxis= {"title":"Population"}
)

fig.show()
```

```python
fig = go.Figure()

trace_0= go.Scatter(
    x= df["Year"],
    y= df["value"],
    mode= "markers+lines"
)

fig.add_trace(trace_0)

fig.update_layout(
    title="USA Population Growth",
    xaxis= {"title":"Year"},
    yaxis= {"title":"Population"}
)


fig.show()
```
