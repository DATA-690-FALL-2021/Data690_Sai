```python
import pandas as pd
import plotly.express as px
```

```python
df= pd.read_csv("/content/wdi_data (4).csv")
df.head()
```

```python
df.sample(10)
```

```python
df_2019= df.query('Year==2019')
df_2019.sample(10)
```

```python
df_2019 = df_2019.query("indicator in(['SP.DYN.LE00.IN','NY.GDP.PCAP.PP.CD'])")
df_2019
```

```python
df_2019.drop(columns=["Unnamed: 0","Year"], inplace= True)
df_2019.head()
```

```python
df_pivot = df_2019.pivot(index=['Country Code','Country Name','Region','Income Group','Lending Type'],columns ='indicator',values = 'value').reset_index()
df_pivot.head()
```

```python
df_pivot.info()
```

```python
fig = px.scatter(
    df_pivot,
    x= "NY.GDP.PCAP.PP.CD",
    y= "SP.DYN.LE00.IN",
    title= "Health vs Wealth",
    labels= {"SP.DYN.LE00.IN":"Life Expectency","NY.GDP.PCAP.PP.CD":"GDP Per Capita Income"},
    color = "Region"

)

fig.show()
```
