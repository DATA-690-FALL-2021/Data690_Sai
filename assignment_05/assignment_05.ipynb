## Defining The Chosen Variables 
## 1 UNITID:
# It defines the Identification Number of the University which is unique


## 2 INSTNM:
# It defines the name of the particular Institution

## 3 City: 
# It defines the place where the University is located

## 4 STABBR:
# This reports the location of the University as well

## 5 ZIP:
# This defines the unique code of a particular region/place. In the above data, it simplifies in locating the universities

## 6 ADM_RATE:
# This defines the rate of admissions of undergrad students to the university divided by number of undergraduates applied.

## 7 UGDS:
# This defines the admission of undergrad students into the university

## 8 TUITIONFEE_IN:
# This defines the cost of fee which include books,extra curricular activities

## 9 MN_EARN_WNE_P10:
# It defines the Mean earnings of students working but not enrolled
